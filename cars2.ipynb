```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
pd.options.display.max_columns = 30
import time
```

```python
url = "https://www.autoscout24.com/lst/porsche/911?sort=year&desc=0&atype=C&ustate=N%2CU&powertype=kw&search_id=65usl0qi6g"
requests.get(url)
html = requests.get(url).content
soup = bs(html)

model = soup.find_all("div",{"class":"ListItem_wrapper__J_a_C"})
model[0].find("a",{"class":"ListItem_title__znV2I Link_link__pjU1l"}).get_text()
model[0].find("p",{"Price_price__WZayw"}).get_text()
model[0].find("span",{"VehicleDetailTable_item__koEV4"}).get_text()
model[0].find("div",{"VehicleDetailTable_container__mUUbY"}).get_text()
```

```python
name = []
price = []
mileage = []
#mileage,first registration,power, - , gearbox, fuel, fuel usage
mass = []

for tech in model:
    #name
    name.append(tech.find("a",{"class":"ListItem_title__znV2I Link_link__pjU1l"}).get_text())
#print(name)
    
    #price
    price.append(tech.find("p",{"Price_price__WZayw"}).get_text())
#print(price)

    #mileage
    mileage.append(tech.find("span",{"VehicleDetailTable_item__koEV4"}).get_text())
#print(mileage)

    #mass
    mass.append(tech.find("div",{"VehicleDetailTable_container__mUUbY"}).get_text())
#print(mass)
```

```python
df = pd.DataFrame({"Name":name,"Mileage": mileage ,"Price":price,"Mileage":mileage, "Mass": mass})
df
```

```python
# Full data scraping from web

name = []
mileage = []
price = []
mass = []


for x in range(1,198):
    url = f"https://www.autoscout24.com/lst/porsche/911?sort=year&desc=0&atype=C&ustate=N%2CU&powertype=kw&search_id=65usl0qi6g&page={x+1}"
    requests.get(url)
    html = requests.get(url).content
    soup = bs(html)
    time.sleep(0.4)

    model = soup.find_all("div",{"class":"ListItem_wrapper__J_a_C"})
    
    for tech in model:
        name.append(tech.find("a",{"class":"ListItem_title__znV2I Link_link__pjU1l"}).get_text())

        price.append(tech.find("p",{"Price_price__WZayw"}).get_text())

        mileage.append(tech.find("span",{"VehicleDetailTable_item__koEV4"}).get_text())

        mass.append(tech.find("div",{"VehicleDetailTable_container__mUUbY"}).get_text())
```

```python
df = pd.DataFrame({"Name": name, "Price": price, "Mileage": mileage, "Mass": mass})
```
